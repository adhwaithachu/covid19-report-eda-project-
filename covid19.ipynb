```python
#Question1:Which countries currently have the highest number of active COVID-19 cases?

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('country_wise.csv')

country_cases = df.groupby('Country/Region')['Active'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(country_cases.index, country_cases.values, color='skyblue', edgecolor='black')
plt.title('Top 10 Countries with Highest Active COVID-19 Cases')
plt.xlabel('Country')
plt.ylabel('Active Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Question2:Which countries have reported the highest number of COVID-19 deaths?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("country_wise.csv")

top = df.sort_values('Deaths', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.histplot(
    data=top,
    x='Country/Region',
    weights='Deaths',
    color='brown',
    bins=10
)

plt.title('Top 10 Countries by COVID-19 Deaths (using histplot)')
plt.xlabel('Country')
plt.ylabel('Deaths')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
#Question3:Which countries have the highest number of confirmed COVID-19 cases?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("country_wise.csv")

top_confirmed = df.sort_values("Confirmed", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=top_confirmed,
    x='Country/Region',
    y='Confirmed',
    s=150,              
    color='red',       
    edgecolor='black'
)

plt.title('Top 10 Countries with Highest Confirmed COVID-19 Cases ')
plt.xlabel('Country')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Question4:Which countries have the highest COVID-19 recovery rate?

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("country_wise.csv")

top_recovered = df.sort_values("Recovered", ascending=False).head(10)

plt.figure(figsize=(8,8))
plt.pie(
    top_recovered["Recovered"],
    labels=top_recovered["Country/Region"],
    autopct="%1.1f%%",
    startangle=140,
    colors=plt.cm.Greens_r(range(10))
)

plt.title("Top 10 Countries Contributing to Global COVID-19 Recoveries")
plt.tight_layout()
plt.show()

```

```python
#Question5:How do confirmed, recovered, and death counts compare among the top 10 countries?

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("country_wise.csv")

top = df.sort_values("Confirmed", ascending=False).head(10)

plt.plot(top["Country/Region"], top["Confirmed"], label="Confirmed", marker='o')
plt.plot(top["Country/Region"], top["Recovered"], label="Recovered", marker='o')
plt.plot(top["Country/Region"], top["Deaths"], label="Deaths", marker='o')

plt.title("COVID-19 Cases Comparison (Top 10 Countries)")
plt.xlabel("Country")
plt.ylabel("Number of Cases")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
